```python
# load data
pheno = PhenologicalData()
#pheno.download_data()
pheno.load_data()
```

```python
# prepare data for plots and analysis
hists = pheno.aggregate_data('count')

mean = pheno.aggregate_data('mean')
median = pheno.aggregate_data('median')
std = pheno.aggregate_data('sd')

# restrict data to official temporal coverage (1951-2019)
# https://opendata.dwd.de/climate_environment/CDC/observations_germany/phenology/annual_reporters/wild/historical/DESCRIPTION_obsgermany-phenology-annual_reporters-wild-historical_en.pdf
mean = mean[(mean.index > 1950) & (mean.index < 2020)]
median = median[(median.index > 1950) & (median.index < 2020)]
std = std[(std.index > 1950) & (std.index < 2020)]
```

```python
# Figure 1: Horizontal Bar Plot
make_horizontal_bar_plot(mean)
```

```python
# Figure 2: Histogram
make_hist_plot(hists)
```

```python
# Figure 3: Time-Series with differences between starting/end point of trend line
make_timeseries_plots(mean, median, std, plot_median=False, plot_median_trend=False)
```

```python
# two-sided t-test of slope parameter, null is zero slope
simple_t_test(mean)
```
